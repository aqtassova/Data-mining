```python
import numpy as np
import pandas as pd

# Загрузите CSV-файл в DataFrame
df = pd.read_csv('data.csv')

# 1. Создать условия взаимодействия
df['Pulse_Maxpulse_Diff'] = df['Maxpulse'] - df['Pulse']
df['Calories_per_minute'] = df['Calories'] / df['Duration']

# 1.2. Полиномиальные характеристики (например, квадрат импульса)
df['Pulse_squared'] = df['Pulse'] ** 2
df['Duration_squared'] = df['Duration'] ** 2

# 3. Функция знания предметной области
df['High_Intensity'] = np.where(df['Maxpulse'] > 150, 1, 0)

# Отобразить новый набор данных с разработанными функциями
df.head()
```
